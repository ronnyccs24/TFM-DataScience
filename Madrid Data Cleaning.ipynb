<div style="text-align: left;">
    <span style="color:#2467C0; font-family: 'Courier New'; font-size: 1.5em;">Nuestra tabla de datos tiene muchas líneas o registros. Para no afectar el funcionamiento de nuestro ordenador, cuando queramos revisar los tipos de datos sólo miraremos pocas líneas a la vez. Ahora bien, en vez de mirar las mismas 2, 5 ó 10 líneas principales cada vez, aplicaremos lo siguiente con lo que conseguiremos ver una muestra aleatoria diferente en nuestro head(). De esta manera, nos aseguramos de que los datos no estén en ningún orden particular cuando probamos el muestreo (como tomar filas de arriba o de abajo) al mezclar aleatoriamente las filas.<br><br></span>
</div>

```python
#Shuffle de las filas de "listing" para que obtengamos una muestra distribuida cuando mostramos el head()

listing = listing.reindex(np.random.permutation(listing.index))
listing.head(3)
```

```python
def transform_price(df, variables=None):
    df = df.copy()
    for variable in variables:
        df[variable] = df[variable].str.strip('$').str.replace(',', '') \
            # Remove dollar signs and thousands separators
        df[variable] = df[variable].astype(np.float64)
    return df


vars_price = ['cleaning_fee', 'price', 'extra_people', 'weekly_price', 'monthly_price' ]
listing = transform_price(listing, vars_price)
```

EL dataset calendar contiene data de los precios para cada piso en alquiler para todo un año y con información de la disponibilidad de los mismos.
<br><br>
Al igual que el dataset anterior, revisamos que los tipos de datos sean los correctos y, si no lo son, convertirlos. 
<br><br>
Nos damos cuenta de que la columna "date", "available" y "price" no tienen el formato que deberían. 
Date debe ser convertido al formato datetime, "price" al igual que el dataset anterior debe cambiarse a float, eliminando previamente el símbolo de dólar y eliminando el punto de separador de miles; y finalmente "available" debe ser una variable boolean.
</span>
</div>

```python
# Cambiamos el formato de las fechas
calendar['date'] = pd.to_datetime(calendar.date, infer_datetime_format=True)
calendar['date'].head(3)
```

```python
# Quitamos el signo de dólar y cualquier separador de miles donde el precio no sea NA
calendar['price'] = calendar.loc[calendar.price.notnull()].price.apply(lambda x: re.sub('[$,]', '', x))
```

```python
# Price se cambia a numérico
calendar['price'] = pd.to_numeric(calendar.price)
calendar['price'].head(3)
```

```python
# La columna 'available' se convierte en boolean
calendar['available'] = calendar['available'].map({'t': True, 'f': False})
calendar['available'].head(3)
```
